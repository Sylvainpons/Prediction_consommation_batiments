```python
from sklearn.preprocessing import OneHotEncoder

# Instantiate OneHotEncoder object with min_frequency parameter
encoder = OneHotEncoder(handle_unknown='ignore', min_frequency=0.5,sparse_output=False)

# Fit and transform the variables
one_encoding_encoded = encoder.fit_transform(one_encoding[['BuildingType', 'PrimaryPropertyType', 'ListOfAllPropertyUseTypes',
       'LargestPropertyUseType', 'SecondLargestPropertyUseType',
       'ThirdLargestPropertyUseType']])

# Generate the column names
columns = np.hstack([encoder.categories_[i].tolist() for i in range(len(encoder.categories_))])
df_building_encoded = pd.DataFrame(one_encoding, columns=columns)

# Concatenate the two dataframes
df_one_enco = pd.concat([one_encoding, one_encoding_encoded], axis=1)
```
